# SF Permits - Data Cleanup

Dataset: https://www.kaggle.com/rtatman/data-cleaning-challenge-handling-missing-values/data

```python
import pandas as pd
import numpy as np
```

```python
sf_permits = pd.read_csv('data/Building_Permits.csv')
np.random.seed(0)
```

```python
sf_permits.describe()
```

```python
sf_permits.sample(5)
```

```python
sf_permits_missing_values_count = sf_permits.isnull().sum()
sf_permits_missing_values_count
```

```python
total_cells = np.product(sf_permits.shape)
total_missing = sf_permits_missing_values_count.sum()

(total_missing/total_cells) * 100
```

**Street Number Suffix** may be genuinely empty because there was no suffix, but **Zipcode** column is missing data. Zipcode data was not recorded.

```python
# What if we dropped all rows in null values?
sf_permits.dropna()
```

```python
# Remove columns that have atleast one null
columns_with_na_dropped = sf_permits.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# How much data did we lose?
print("Columns in original dataset: %d \n" % (sf_permits.shape[1]))
print("Columns in na's dropped: %d \n" % (columns_with_na_dropped.shape[1]))
```
