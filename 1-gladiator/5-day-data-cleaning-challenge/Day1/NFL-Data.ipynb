Dataset: https://www.kaggle.com/maxhorowitz/nflplaybyplay2009to2016/data



```python
import pandas as pd
import numpy as np
```

```python
nfl_data = pd.read_csv('data/NFL Play by Play 2009-2017 (v4).csv')

np.random.seed(0)
```

```python
nfl_data.describe()
```

```python
nfl_data.sample(5)
```

```python
nfl_data_missing_values_count = nfl_data.isnull().sum()
nfl_data_missing_values_count
```

```python
total_cells = np.product(nfl_data.shape)
total_missing = nfl_data_missing_values_count.sum()

(total_missing/total_cells) * 100
```

```python
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# Just how much data did we lose?
print("Columns in original dataset: %d \n" % (nfl_data.shape[1]))
print("Columns with na's dropped: %d \n" % (columns_with_na_dropped.shape[1]))
```

```python
subset_nfl_data = nfl_data.loc[:, 'EPA':'Season'].head()
subset_nfl_data
```

```python
subset_nfl_data.fillna(0)
```

```python
subset_nfl_data.fillna(method='bfill', axis=0).fillna(0)
```
